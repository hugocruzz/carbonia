```python
# Import necessary libraries
import os
import pandas as pd
from embeddings import embed
from match import *

# Set your OpenAI API key
# Example 1: Embed a string
string_to_embed = "This is a test string."
embedded_string = embed(string_to_embed)
print("Embedded String:", embedded_string)

# Example 2: Embed a list of strings
list_of_strings_to_embed = ["This is the first test string.", "This is the second test string."]
embedded_list = embed(list_of_strings_to_embed)
print("Embedded List of Strings:", embedded_list)

# Example 3: Embed a pandas DataFrame
data = {
    "text": ["This is the first test string.", "This is the second test string.", "a microscope"]
}
df_to_embed = pd.DataFrame(data)
embedded_df = embed(df_to_embed)
print("Embedded DataFrame:")
print(embedded_df)

# Example 4: Match embedded DataFrame with a target DataFrame
target_data = {
    "text": ["First target string.", "Second target string.", "a string", "some microscope"]
}
df_target = pd.DataFrame(target_data)

# Assuming the target DataFrame also needs to be embedded
embedded_target_df = embed(df_target)

```

```python
# Match the embedded DataFrame with the target DataFrame
from krbon.match import match_datasets

matched_df = match_datasets(embedded_df.copy(), embedded_target_df.copy(), top_n=1, gpt_model="gpt-4o-mini", api_key=os.getenv("OPENAI_API_KEY"))
print("Matched DataFrame:")
print(matched_df["combined"])
print(matched_df["combined_target"])
```
